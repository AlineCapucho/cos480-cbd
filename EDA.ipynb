# Exploring The Dataset

```python
data = pd.read_csv('./dataset/Coffee Shop Sales.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
g= sns.catplot(
    data=data, y="product_type", hue="product_category", kind="count",
    palette="pastel", edgecolor=".6",
)
g.set_xlabels('Amount sold')
g.set_ylabels('Produt')
g.set_titles('Sales per product')
g._legend.set_title("Product category")
plt.show()
```

```python
g = sns.catplot(data=data, x="product_category", kind="count")
g.set_xlabels('Product category')
g.set_ylabels('Count')
g.set_titles('Sales per product category')
g.set_xticklabels(rotation=90)
plt.show()
```
